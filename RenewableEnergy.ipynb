```python
#Make a Loop that Creates an Model for Solar Energy Consumption by Month
```

```python
#Transform Monthly Solar Energy Consumption from 2005 - 2023 into a data array

x = np.array(list(range(0, 19)))

n_months = 12
n_years = len(x)

y = np.zeros((n_years, n_months))

for i in range(0,12):

    y[:,i] = list(df_month[(df_month['Year']>2004) & (df_month['Month']==i+1)]['Solar Energy'])
```

```python
#Exponential Curve Fit

from scipy.optimize import curve_fit

# Define exponential function
def exp_func(x, a, b, c):
    return a * np.exp(b * x) + c

#Create list of coefficient fit parameters for each month
a_list = []
b_list = []
c_list = []

for i in range(0,12):

    params, cov = curve_fit(exp_func, x, y[:,i], p0=(1, 0.1,0),maxfev=10000)  # p0 = initial guess (a=1, b=1, c=1)
    a, b, c = params

    a_list.append(a)
    b_list.append(b)
    c_list.append(c)
```

```python
#Visually assess model fit for all twelve months
```

```python
fig, axes = plt.subplots(3, 4, figsize=(12, 10))

axes[0,0].plot(x+2005, exp_func(x, a_list[0], b_list[0], c_list[0]), color=[0.0660, 0.4430, 0.7450], label='Model')
sns.scatterplot(x='Year',y='Solar Energy',data=df_month[(df_month['Month']==1) & (df_month['Year']>2004)],color=[0.0660, 0.4430, 0.7450], label='Data',ax=axes[0,0])
axes[0,0].set_title('January')
axes[0,0].set_ylabel('Trillion BTUs')

axes[0,1].plot(x+2005, exp_func(x, a_list[1], b_list[1], c_list[1]), color=[0.0660, 0.4430, 0.7450], label='Model')
sns.scatterplot(x='Year',y='Solar Energy',data=df_month[(df_month['Month']==2) & (df_month['Year']>2004)],color=[0.0660, 0.4430, 0.7450], label='Data',ax=axes[0,1])
axes[0,1].set_title('February')
axes[0,1].set_ylabel('Trillion BTUs')

axes[0,2].plot(x+2005, exp_func(x, a_list[2], b_list[2], c_list[2]), color=[0.0660, 0.4430, 0.7450], label='Model')
sns.scatterplot(x='Year',y='Solar Energy',data=df_month[(df_month['Month']==3) & (df_month['Year']>2004)],color=[0.0660, 0.4430, 0.7450], label='Data',ax=axes[0,2])
axes[0,2].set_title('March')
axes[0,2].set_ylabel('Trillion BTUs')

axes[0,3].plot(x+2005, exp_func(x, a_list[3], b_list[3], c_list[3]), color=[0.0660, 0.4430, 0.7450], label='Model')
sns.scatterplot(x='Year',y='Solar Energy',data=df_month[(df_month['Month']==4) & (df_month['Year']>2004)],color=[0.0660, 0.4430, 0.7450], label='Data',ax=axes[0,3])
axes[0,3].set_title('April')
axes[0,3].set_ylabel('Trillion BTUs')

axes[1,0].plot(x+2005, exp_func(x, a_list[4], b_list[4], c_list[4]), color=[0.0660, 0.4430, 0.7450], label='Model')
sns.scatterplot(x='Year',y='Solar Energy',data=df_month[(df_month['Month']==5) & (df_month['Year']>2004)],color=[0.0660, 0.4430, 0.7450], label='Data',ax=axes[1,0])
axes[1,0].set_title('May')
axes[1,0].set_ylabel('Trillion BTUs')

axes[1,1].plot(x+2005, exp_func(x, a_list[5], b_list[5], c_list[5]), color=[0.0660, 0.4430, 0.7450], label='Model')
sns.scatterplot(x='Year',y='Solar Energy',data=df_month[(df_month['Month']==6) & (df_month['Year']>2004)],color=[0.0660, 0.4430, 0.7450], label='Data',ax=axes[1,1])
axes[1,1].set_title('June')
axes[1,1].set_ylabel('Trillion BTUs')

axes[1,2].plot(x+2005, exp_func(x, a_list[6], b_list[6], c_list[6]), color=[0.0660, 0.4430, 0.7450], label='Model')
sns.scatterplot(x='Year',y='Solar Energy',data=df_month[(df_month['Month']==7) & (df_month['Year']>2004)],color=[0.0660, 0.4430, 0.7450], label='Data',ax=axes[1,2])
axes[1,2].set_title('July')
axes[1,2].set_ylabel('Trillion BTUs')

axes[1,3].plot(x+2005, exp_func(x, a_list[7], b_list[7], c_list[7]), color=[0.0660, 0.4430, 0.7450], label='Model')
sns.scatterplot(x='Year',y='Solar Energy',data=df_month[(df_month['Month']==8) & (df_month['Year']>2004)],color=[0.0660, 0.4430, 0.7450], label='Data',ax=axes[1,3])
axes[1,3].set_title('August')
axes[1,3].set_ylabel('Trillion BTUs')

axes[2,0].plot(x+2005, exp_func(x, a_list[8], b_list[8], c_list[8]), color=[0.0660, 0.4430, 0.7450], label='Model')
sns.scatterplot(x='Year',y='Solar Energy',data=df_month[(df_month['Month']==9) & (df_month['Year']>2004)],color=[0.0660, 0.4430, 0.7450], label='Data',ax=axes[2,0])
axes[2,0].set_title('September')
axes[2,0].set_ylabel('Trillion BTUs')

axes[2,1].plot(x+2005, exp_func(x, a_list[9], b_list[9], c_list[9]), color=[0.0660, 0.4430, 0.7450], label='Model')
sns.scatterplot(x='Year',y='Solar Energy',data=df_month[(df_month['Month']==10) & (df_month['Year']>2004)],color=[0.0660, 0.4430, 0.7450], label='Data',ax=axes[2,1])
axes[2,1].set_title('October')
axes[2,1].set_ylabel('Trillion BTUs')

axes[2,2].plot(x+2005, exp_func(x, a_list[10], b_list[10], c_list[10]), color=[0.0660, 0.4430, 0.7450], label='Model')
sns.scatterplot(x='Year',y='Solar Energy',data=df_month[(df_month['Month']==11) & (df_month['Year']>2004)],color=[0.0660, 0.4430, 0.7450], label='Data',ax=axes[2,2])
axes[2,2].set_title('November')
axes[2,2].set_ylabel('Trillion BTUs')

axes[2,3].plot(x+2005, exp_func(x, a_list[11], b_list[11], c_list[11]), color=[0.0660, 0.4430, 0.7450], label='Model')
sns.scatterplot(x='Year',y='Solar Energy',data=df_month[(df_month['Month']==12) & (df_month['Year']>2004)],color=[0.0660, 0.4430, 0.7450], label='Data',ax=axes[2,3])
axes[2,3].set_title('December')
axes[2,3].set_ylabel('Trillion BTUs')


plt.tight_layout()

plt.savefig("solar_energy_3.png", dpi=300, bbox_inches='tight') 
```

```python
#Project model five years into the future
```

```python
a_list = np.array(a_list)
b_list = np.array(b_list)
c_list = np.array(c_list)

x_future = np.array(list(range(19, 24)))
y_future = exp_func(x_future[:, None], a_list[None,:], b_list[None,:], c_list[None,:])
```

```python
#Convert NumPy array into DataFrame

years = x_future + 2005
months = list(range(1, 13))

df_wide = pd.DataFrame(y_future, index=years, columns=months)
df_wide.index.name = 'Year'
df_wide.columns.name = 'Month'

#Change Shape of DataFrame to match formatting for df_month
df_future = df_wide.reset_index().melt(
    id_vars='Year', 
    var_name='Month', 
    value_name='Solar Energy'
)

# Sort by Year first, then Month
df_future = df_future.sort_values(by=['Year', 'Month']).reset_index(drop=True)
df_future.head()

#Create Date Column 
df_future['Date'] = df_future.apply(lambda x: datetime(int(x['Year']), int(x['Month']), 1), axis=1)

df_future.tail()
```

```python
#Plot Date vs. Solar Energy Consumption with new Model Predictions

plt.figure(figsize=(8, 6))
ax = sns.lineplot(x='Date',y='Solar Energy',data=df_month[df_month['Year']>1990],label='Data')
ax = sns.lineplot(x='Date',y='Solar Energy',data=df_future,label='Model Forecast')

plt.legend(frameon=False,fontsize=14)
plt.title('Solar Energy Consumption',fontsize=16)
plt.ylabel('Trillion BTUs',fontsize=14)
plt.xlabel('Year',fontsize=14)
ax.tick_params(axis='both',labelsize=14)
ax.set_xlim([dt.datetime(1990, 1, 1), dt.datetime(2029, 1, 1)])
plt.tight_layout()

plt.savefig("solar_energy_4.png", dpi=300, bbox_inches='tight') 
```
